```python
from rdkit import Chem
from rdkit.Chem import rdMolAlign

# Absolute paths to ligand PDB files
crystal = Chem.MolFromPDBFile(
    r"D:\MeninQSAR\7UJ4validation\Crystal_ligand.pdb",
    removeHs=True
)

docked = Chem.MolFromPDBFile(
    r"D:\MeninQSAR\7UJ4validation\re-docked\Redocked_ligand.pdb",
    removeHs=True
)

if crystal is None or docked is None:
    raise ValueError("Failed to read PDB files. Check PDB formatting.")

# Sanity check
if crystal.GetNumAtoms() != docked.GetNumAtoms():
    raise ValueError("Ligands do not have the same number of atoms.")

#  RMSD (accounts for symmetry)
rmsd = rdMolAlign.GetBestRMS(crystal, docked)

print(f"Heavy-atom RMSD between crystal and docked pose: {rmsd:.3f} Å")
```
